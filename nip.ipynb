```python
import nltk
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')


# init stemmer
porter_stemmer=PorterStemmer()
# stem connect variations
words=["connect","connected","connection","connections","connects"]
stemmed_words=[porter_stemmer.stem(word=word) for word in words]

print(words)
print(stemmed_words)

# init lemmatizer
lemmatizer = WordNetLemmatizer() 
words=["trouble","troubling","troubled","troubles","dogs", "cats"]
lemmatized_words=[lemmatizer.lemmatize(word=word,pos='v') for word in words]
print(words, lemmatized_words)
```

```python
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer

nltk.download('wordnet')
nltk.download('punkt')

mytext = """CDW Corporation is a leading provider of technology products and services for businesses, governments, and educational institutions. Founded in 1984, CDW has grown into a Fortune 500 company, offering a wide range of hardware, software, and integrated IT solutions. Its services include cloud computing, cybersecurity, digital workspace solutions, and data center optimization."""

words = nltk.word_tokenize(mytext)
print(words)


processed_text = (mytext)
print("".join(processed_text))

porter_stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()


word=["connect","connected","connection","connections","connects"]
stemmed_word=[porter_stemmer.stem(word=word) for word in words]

word=["trouble","troubling","troubled","troubles","dogs", "cats"]
lemmatized_word=[lemmatizer.lemmatize(word=word,pos='v') for word in words]

print(word, lemmatized_word)


stemmed_words = [porter_stemmer.stem(word) for word in words]
lemmatized_words = [lemmatizer.lemmatize(word, pos='v') for word in words]

for original, stemmed, lemmatized in zip(words, stemmed_words, lemmatized_words):
    print(f"{original} {stemmed} {lemmatized}")
```
